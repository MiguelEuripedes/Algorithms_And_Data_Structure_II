```python
import networkx as nx
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class GraphHandler:
    def __init__(self, input_file, directed=False):
        if directed:
            self.graph = nx.DiGraph()
        else:
            self.graph = nx.Graph()
        with open(input_file, "r") as infile:
            for line in infile:
                line = line.strip()
                if line.startswith("#") or not line:
                    continue
                nodes = line.split()  
                if len(nodes) == 2:
                    source, target = nodes
                    self.graph.add_edge(source, target)

    def create_degree_vs_mean_neighbor_degree_plot(self, xlim=None):
        degree, avg_neighbor_degree = zip(*nx.average_degree_connectivity(self.graph).items())
        degree = list(degree)
        avg_neighbor_degree = list(avg_neighbor_degree)

        plt.figure(figsize=(10, 6))
        sns.regplot(x=degree, y=avg_neighbor_degree, scatter_kws={"alpha": 0.5})
        plt.title("Assortatividade em Relação ao Grau dos Nós")
        plt.grid(True)

        if xlim:
            plt.xlim(xlim) 

        plt.show()
```

```python
handledPhiliRoadGraph = GraphHandler('/content/roadNet-PA.txt')
```

```python
print("Numero de edges: ", handledPhiliRoadGraph.graph.number_of_edges())
print("Numero de nodes: ", handledPhiliRoadGraph.graph.number_of_nodes())
```

```python
handledPhiliRoadGraph.create_degree_vs_mean_neighbor_degree_plot()
```
